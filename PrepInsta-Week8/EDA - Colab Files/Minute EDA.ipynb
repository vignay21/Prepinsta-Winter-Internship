### 7. Outliers Detection

Identify the outliers in relevant columns.

```python
# Define the z-score threshold for outlier detection
z_score_threshold = 3

# Create an empty dictionary to store outliers for each column
all_outliers = {}

# Iterate through each column in the DataFrame
for column in df.columns:
    # Check if the column contains numerical data
    if pd.api.types.is_numeric_dtype(df[column]):
        # Identify and store outliers for the current column
        outliers = df[column][np.abs(stats.zscore(df[column])) > z_score_threshold]
        all_outliers[column] = outliers

# Display the dictionary of outliers for each column
for column, outliers in all_outliers.items():
    print(f"Outliers in {column}:\n{outliers}\n")
```

**Observation:**<br>
Determining the outliers for every column in Dataframe

### 9. Total Sleep by Every User

```python
# Filter rows where 'sleep' is not 0
sleep_data = df[df['sleep'] != 0]

# Calculate total sleep duration for each 'id'
total_sleep_by_id = sleep_data.groupby('id')['sleep'].sum().reset_index()

# Sort the DataFrame in ascending order based on total sleep duration
total_sleep_by_id_sorted = total_sleep_by_id.sort_values(by='sleep', ascending=True)

# Display the sorted results
print("Total sleep duration by id (sorted in ascending order):")
print(total_sleep_by_id_sorted)
```

### 1. Heatmap

Visualizing the Steps, METs, Calories, Intensity and Sleep with Heat Map
